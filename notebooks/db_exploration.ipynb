```python
import duckdb

con = duckdb.connect("../data/bgg_db.duckdb")
```

# DB Overview

```python
# Number of games
con.sql("SELECT COUNT(1) FROM bgg.games;")
```

```python
# Number of comments
con.sql("SELECT COUNT(1) FROM bgg.comments;")
```

```python
# Distribution of comment ratings
con.sql("SELECT bgg_id, AVG(rating) FROM bgg.comments GROUP BY bgg_id")
```

# Finding Existing Games

Check if an existing game ID exists in the table

```python
GAME_ID = 291453
con.sql(f"SELECT * FROM bgg.games WHERE id = {GAME_ID}")
```

```python
con.sql(f"SELECT COUNT(1) FROM bgg.comments WHERE bgg_id = {GAME_ID}")
```
